```python
import pandas as pd
import geopandas # we'll use this to read shapefiles
from shapely.geometry import Point

pd.set_option('display.max_rows', 1500)
```

```python
# the City_Limit directory contains some shapefiles.
# downloaded from here: http://data-cityofmadison.opendata.arcgis.com/datasets/city-limit
madison = geopandas.read_file("City_Limit")
```

```python
# downloaded from here: http://data-cityofmadison.opendata.arcgis.com/datasets/c46082b091a941f8b2ded1dd115a1a05_8
lakes = geopandas.read_file("Lakes_and_Rivers")
bus = geopandas.read_file("Metro_Transit_Bus_Routes")
rtbus = geopandas.read_file("Bus_Rapid_Transit_System_Lines")
```

```python
import matplotlib.pyplot as plt

ax = madison.plot(color="#303030", figsize=(15,15))

ax.set_facecolor("#212121")
lakes.plot(color="#485155", ax=ax)
bus.plot(color="#A0A0A0", ax=ax)
rtbus.plot(color="#58A1BD",ax=ax)

figure = plt.gcf()
figure.savefig('map', dpi=600)



```

```python
ax = madison.plot(color="#1A1A1A", figsize=(15,15))
ax.set_facecolor("#161616")

lakes.plot(color="#232628", ax=ax)
bus.plot(color="#4D4D4D", ax = ax)
rtbus.plot(color="#2B4E5E",ax=ax)

figure = plt.gcf()
figure.savefig('map-2', dpi=600)
```

```python
ax = bus.plot(color="#303335", figsize = (10,10))
ax.set_facecolor("#383C3F")

figure = plt.gcf()
figure.savefig('map-mini', dpi=200)
```

```python
ax = bus.plot(color="#282828", figsize = (15,15))
rtbus.plot(color="#2B4E5E",ax=ax)
ax.set_facecolor("#161616")

figure = plt.gcf()
figure.savefig('map-simple', dpi=600)
```
